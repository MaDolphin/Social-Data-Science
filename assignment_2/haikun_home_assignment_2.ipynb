### Task 1: Causal Inference (1 pt)

Suppose you conduct a study to evaluate the effect of a new procedure for coronary bypass surgery that is supposed to help with the postoperative healing process. Since the new procedure is more risky than the old one, it is rarely performed on patients who are over 80 years old. However, there is also hardly any data on under 80-year olds that have taken the old treatment. 

You can find the data from the study in "bypass.csv", where _stay_ is the length of the hospital stay after surgery, _age_ is the age of the patient, and _new_ is the binary indicator variable specifying whether the new surgical procedure was used. Additionally, there is a column _severity_ which quantifies the severity of preoperative diseases.

Perform a regression analysis to draw a causal inference whether the new surgery method shortens the length of the postoperative hospital stay. Can you observe a significant effect? Would you generally recommend to apply the new surgery method?
Provide a thorough explanation of your answers and the steps in your analysis!

```python
data_bypass = pd.read_csv("bypass.csv")
df_bypass = pd.DataFrame(data_bypass)
df_bypass = data_bypass[data_bypass['age'] <= 80] # age <= 80
df_bypass_old = df_bypass[data_bypass['new'] == 0] # old surgery
df_bypass_new = df_bypass[data_bypass['new'] == 1] # new surgery

y_new = df_bypass_new.stay
x_new = sm.add_constant(df_bypass_new.severity)
model_new = sm.OLS(y_new,x_new)
results_new = model_new.fit()
# print(results_new.summary())
print("new regression params = ", results_new.params)

y_old = df_bypass_old.stay
x_old = sm.add_constant(df_bypass_old.severity)
model_old = sm.OLS(y_old,x_old)
results_old = model_old.fit()
# print(results_old.summary())
print("old regression params = ", results_old.params)

y_new_fitted = results_new.fittedvalues
y_old_fitted = results_old.fittedvalues
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(df_bypass_new.severity, y_new, 'o', label='new_surgery data')
ax.plot(df_bypass_old.severity, y_old, 'x', label='old_surgery data')
new_OLS = ax.plot(df_bypass_new.severity, y_new_fitted, 'r--.',label='new_surgery OLS')
old_OLS = ax.plot(df_bypass_old.severity, y_old_fitted, 'g--.',label='old_surgery OLS')
plt.xlabel('Severity which quantifies the severity of preoperative diseases')
plt.ylabel('The length of the postoperative hospital stay')
plt.title('Regression analysis between the new surgery method and the length of the postoperative hospital stay')
ax.legend()
```

### Task 2: Naive Bayes on Continuous Data (1 pt)

In lecture we have introduced the _Naive Bayes_ classifier, in which for each feature, we compute the empirical probabilities 
$ P(x_i|y) $ to form our predictions. For discrete data, we have learned in lecture that these probabilities are estimated from the raw class-conditional feature counts. For continuous features however, this approach obviously does not make sense.
Instead, for continuous features, one usually makes the assumption that these  are normally distributed. Thus for continuous features, one usually models the class-conditional probabilities via 
$$ P(X_i = x| y = C_k) = \frac{1}{\sqrt{2\pi \sigma_{ik}^2}} \exp\left(-\frac{x_i-\mu_{ik}}{2\sigma_{ik}^2}\right)$$
where $\mu_{ik}$ and $\sigma_{ik}$ are the mean and variance of the features $X_i$ that are classed as $c_k$.
A Naive Bayes classifier for which all features are assumed to be continuous and normally distribited is also called _Gaussian Naive Bayes_ classifier.

Implement a class that performs Naive Bayes classification on both discrete and continuous data, using the class and function signatures and specifications in the cell below, which are designed to be similar to the model classes in sklearn. 
When passing your data to the fit() function, by default your classifier should treat all columns as continuous, i.e. normally distributed as explained above. Discrete features have to be specified upon fitting, by passing a list of integers specifying the indices of the columns that are to be treated as discrete features.

Note: You may use any function from the numpy library, but __NO__ function from the scikit-learn library. Further, you may assume that all input data is valid, i.e. you do not need to check whether for instance the feature matrix X has as many rows as the class vector $y$.

```python
class NaiveBayes:
    
    # Bayes Rule, P(Y|X) = P(X|Y)*P(Y)/P(X)
    # P(X|Y) - Gaussian distribution (calculate_gaussian_probability)
    # P(Y) - Prior (calculate_prior_probability)
    # P(X) - Normalization Factor (ignored)
    # posterior P(Y|X) = log(P(X|Y)) + log(P(Y)) (calculate_posterior_probability)
    # we want to get the largest P(Y|X)
    
    def __init__(self):
        # your code here
        self.classes = None
        self.X = None
        self.y = None
        # to store mean and var about gaussian_distribution
        # each feature has his own parameters of gaussian_distribution
        self.parameters = []
     
    
    # find each value of mean and var for each feature in order to calculate_gaussian_probability
    # self.parameters = []
    # fit your model, nothing is returned
    # X: two dimensional numpy array describing the feature matrix to train on
    # y: one dimensional numpy array or list representing the class vector used in training
    # discrete columns: optional list of indices describing the coumns in X that are to be treated as numerical.
    def fit(self, X, y, discrete_columns = None):
        # your code here
        self.X = X
        self.y = y
        self.classes = np.unique(y)
        for i in range (len(self.classes)):
            feature_class = self.classes[i]
            x_in_class = X[np.where(y == feature_class)] # choose the data which class is equal to be aboved
            self.parameters.append([])
            for j in range(len(x_in_class[0, :])): # choose each column and calculate the mean and variance
                col = x_in_class[:, j]
                parameters = {} # create a dict for each mean and variance of column
                parameters["mean"] = col.mean()
                parameters["var"] = col.var()
                self.parameters[i].append(parameters)
    
    
    # calculate_gaussian_probability to caluculate likelihood
    def calculate_gaussian_probability(self, mean, var, x):
        coeff = (1.0 / (np.sqrt((2.0 * np.pi) * var)))
        exp = np.exp(-np.power((x - mean),2)/ (2 * var))
        return coeff * exp
    
    
    # calculate prior
    def calculate_prior_probability(self, feature_class):
        x_in_class = self.X[np.where(self.y == feature_class)]
        num_samples_for_feature_class = x_in_class.shape[0]
        num_samples = self.X.shape[0]
        return num_samples_for_feature_class / num_samples

    
    # calculate likelihood
    def calculate_likelihood(self,params,sample_feature):
        mean = params["mean"]
        var = params["var"]
        return self.calculate_gaussian_probability(mean, var, sample_feature)

    
    # calculate posterior
    # posterior P(Y|X) = log(P(X|Y)) + log(P(Y))
    def calculate_posterior_probability(self,sample):
        posteriors = []
        
        for i in range(len(self.classes)):
            feature_class = self.classes[i]
            prior = self.calculate_prior_probability(feature_class)
            posterior = np.log(prior)
            
            # probability = P(Y)*P(x1|Y)*P(x2|Y)*...*P(xN|Y)
            for j, params in enumerate(self.parameters[i]):
                sample_feature = sample[j]
                likelihood = self.calculate_likelihood(params,sample_feature)
                # In BayesModel is conditonal independent，P(x1,x2,x3|Y) = P(x1|Y)*P(x2|Y)*P(x3|Y)
                # posterior P(Y|X) = log(P(X|Y)) + log(P(Y))
                posterior = posterior + np.log(likelihood)

            posteriors.append(posterior)
        
        index_max = np.argmax(posteriors)
        return self.classes[index_max]

            
    # predict function: use the previously trained model to make a prediction
    # -> return a vector y_hat of predicted classes.
    # X: two dimensional numpy array describing the feature matrix to predict on
    def predict(self, X):
        # your code here
        y_hat = []
        for sample in X:
            y = self.calculate_posterior_probability(sample)
            y_hat.append(y)
        return np.array(y_hat)
    
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize
from sklearn.metrics import accuracy_score
from sklearn import datasets

data = datasets.load_iris()
X = normalize(data.data, norm = 'l2')
y = data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
clf = NaiveBayes()
clf.fit(X_train, y_train)
y_pred = np.array(clf.predict(X_test))

print ("Accuracy:", accuracy_score(y_test, y_pred))
```
